```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
query = "SELECT ConvertedComp FROM master"
df = pd.read_sql_query(query,conn)

plt.hist(df['ConvertedComp'].dropna(),bins=20)
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
query = "SELECT Age FROM master"
df = pd.read_sql_query(query, conn)
plt.boxplot(df['Age'].dropna())
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Extract Age and WorkWeekHrs data
query = "SELECT Age, WorkWeekHrs FROM master"
df = pd.read_sql_query(query, conn)

# Plot scatter plot
plt.scatter(df['Age'], df['WorkWeekHrs'])
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Extract WorkWeekHrs, CodeRevHrs, and Age data
query = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"
df = pd.read_sql_query(query, conn)

# Plot bubble plot
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'].fillna(0)*10, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs with Age as size')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# Replace 'DatabaseDesireNextYear' with the actual column name if different
QUERY = """
SELECT PurchaseWhat as DatabaseDesireNextYear, COUNT(*) as Count
FROM master
GROUP BY PurchaseWhat
ORDER BY Count DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY, conn)
plt.figure(figsize=(10, 7))
plt.pie(df['Count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
AND WorkWeekHrs IS NOT NULL
AND CodeRevHrs IS NOT NULL
"""

df = pd.read_sql_query(QUERY, conn)
# Calculate median WorkWeekHrs and CodeRevHrs
median_work_week_hours = df.groupby('Age')['WorkWeekHrs'].median()
median_code_rev_hours = df.groupby('Age')['CodeRevHrs'].median()
import numpy as np

# Create a stacked chart
ages = median_work_week_hours.index
work_week_hours = median_work_week_hours.values
code_rev_hours = median_code_rev_hours.values

plt.figure(figsize=(10, 7))
plt.bar(ages, work_week_hours, label='Median WorkWeekHrs')
plt.bar(ages, code_rev_hours, bottom=work_week_hours, label='Median CodeRevHrs')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.legend()
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
AND ConvertedComp IS NOT NULL
"""

df = pd.read_sql_query(QUERY, conn)

median_converted_comp = df.groupby('Age')['ConvertedComp'].median()

plt.figure(figsize=(10, 7))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""

df = pd.read_sql_query(QUERY, conn)
# Create a horizontal bar chart
plt.figure(figsize=(10, 7))
plt.barh(df['MainBranch'], df['Count'], color='skyblue')
plt.xlabel('Count')
plt.title('Main Branch Distribution')
plt.show()


```
